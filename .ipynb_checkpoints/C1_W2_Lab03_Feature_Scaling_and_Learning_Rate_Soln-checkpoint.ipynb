## Gradient Descent com Múltiplas Variáveis

Aqui estão as equações que você desenvolveu no último laboratório sobre o Gradiente Descendente para múltiplas variáveis:

$$\begin{align*} \text{repetir}&\text{ até convergência:} \; \lbrace \newline\;
& w_j := w_j -  \alpha \frac{\partial J(\mathbf{w},b)}{\partial w_j} \tag{1}  \; & \text{para j = 0..n-1}\newline
&b\ \ := b -  \alpha \frac{\partial J(\mathbf{w},b)}{\partial b}  \newline \rbrace
\end{align*}$$

onde, n é o número de características, os parâmetros $w_j$, $b$, são atualizados simultaneamente e onde  

$$
\begin{align}
\frac{\partial J(\mathbf{w},b)}{\partial w_j}  &= \frac{1}{m} \sum\limits_{i = 0}^{m-1} (f_{\mathbf{w},b}(\mathbf{x}^{(i)}) - y^{(i)})x_{j}^{(i)} \tag{2}  \\
\frac{\partial J(\mathbf{w},b)}{\partial b}  &= \frac{1}{m} \sum\limits_{i = 0}^{m-1} (f_{\mathbf{w},b}(\mathbf{x}^{(i)}) - y^{(i)}) \tag{3}
\end{align}
$$

* m é o número de exemplos de treinamento no conjunto de dados.
* $f_{\mathbf{w},b}(\mathbf{x}^{(i)})$ é a predição do modelo, enquanto $y^{(i)}$ é o valor alvo.

Por que isso acontece? Existe algo que podemos melhorar? Veja abaixo:
<figure>
    <center> <img src="./images/C1_W2_Lab06_scale.PNG"></center>
</figure>   

A figura acima mostra por que os $w$ são atualizados de forma desigual. 
- $\alpha$ é compartilhado por todas as atualizações de parâmetros ($w$ e $b$).
- o termo de erro comum é multiplicado pelas características para os $w$ (não para $b$).
- as características variam significativamente em magnitude, fazendo com que algumas características sejam atualizadas muito mais rapidamente do que outras. Neste caso, $w_0$ é multiplicado por 'tamanho (sqft)', que geralmente é > 1000, enquanto $w_1$ é multiplicado pelo 'número de quartos', que geralmente é 2-4. 
    
A solução é o Redimensionamento de Características.


As aulas discutiram três técnicas diferentes: 
- Redimensionamento de características, que essencialmente divide cada característica positiva pelo seu valor máximo, ou de forma mais geral, redimensiona cada característica pelo seu valor mínimo e máximo usando (x-min)/(max-min). Ambas as formas normalizam as características para o intervalo de -1 a 1, onde o primeiro método funciona para características positivas, é simples e serve bem para o exemplo da aula, e o segundo método funciona para quaisquer características.
- Normalização pela média: $x_i := \dfrac{x_i - \mu_i}{max - min} $ 
- Normalização pelo Z-score, que exploraremos a seguir.

### Normalização pelo Z-score
Após a normalização pelo Z-score, todas as características terão uma média de 0 e um desvio padrão de 1.

Para implementar a normalização pelo Z-score, ajuste seus valores de entrada conforme mostrado na seguinte fórmula:
$$x^{(i)}_j = \dfrac{x^{(i)}_j - \mu_j}{\sigma_j} \tag{4}$$ 
onde $j$ seleciona uma característica ou uma coluna na matriz $\mathbf{X}$. $µ_j$ é a média de todos os valores para a característica (j) e $\sigma_j$ é o desvio padrão da característica (j).
$$
\begin{align}
\mu_j &= \frac{1}{m} \sum_{i=0}^{m-1} x^{(i)}_j \tag{5}\\
\sigma^2_j &= \frac{1}{m} \sum_{i=0}^{m-1} (x^{(i)}_j - \mu_j)^2  \tag{6}
\end{align}
$$

>**Nota de Implementação:** Ao normalizar as características, é importante armazenar os valores usados para a normalização - o valor médio e o desvio padrão usados para os cálculos. Após aprender os parâmetros do modelo, muitas vezes queremos prever os preços de casas que ainda não vimos antes. Dado um novo valor de x (área da sala de estar e número de quartos), devemos primeiro normalizar x usando a média e o desvio padrão que tínhamos previamente calculado a partir do conjunto de treinamento.

**Implementação**

```python
def zscore_normalize_features(X):
    """
    Calcula X normalizado pelo z-score por coluna
    
    Argumentos:
      X (ndarray (m,n))     : dados de entrada, m exemplos, n características
      
    Retornos:
      X_norm (ndarray (m,n)): entrada normalizada por coluna
      mu (ndarray (n,))     : média de cada característica
      sigma (ndarray (n,))  : desvio padrão de cada característica
    """
    # encontra a média de cada coluna/característica
    mu     = np.mean(X, axis=0)                 # mu terá formato (n,)
    # encontra o desvio padrão de cada coluna/característica
    sigma  = np.std(X, axis=0)                  # sigma terá formato (n,)
    # subtrai, elemento a elemento, a média daquela coluna de cada exemplo e divide pelo desvio padrão daquela coluna
    X_norm = (X - mu) / sigma      

    return (X_norm, mu, sigma)
 
# verificar nosso trabalho
# from sklearn.preprocessing import scale
# scale(X_orig, axis=0, with_mean=True, with_std=True, copy=True)
```

Vamos analisar as etapas envolvidas na normalização pelo Z-score. O gráfico abaixo mostra a transformação passo a passo.

```python
mu     = np.mean(X_train, axis=0)   
sigma  = np.std(X_train, axis=0) 
X_mean = (X_train - mu)
X_norm = (X_train - mu) / sigma      

fig, ax = plt.subplots(1, 3, figsize=(12, 3))

# Plot original data (unnormalized)
ax[0].scatter(X_train[:, 0], X_train[:, 3])
ax[0].set_xlabel(X_features[0])
ax[0].set_ylabel(X_features[3])
ax[0].set_title("unnormalized")
ax[0].axis('equal')

# Plot data after subtracting the mean
ax[1].scatter(X_mean[:, 0], X_mean[:, 3])
ax[1].set_xlabel(X_features[0])
ax[0].set_ylabel(X_features[3])
ax[1].set_title(r"X - $\mu$")
ax[1].axis('equal')

# Plot data after Z-score normalization
ax[2].scatter(X_norm[:, 0], X_norm[:, 3])
ax[2].set_xlabel(X_features[0])
ax[0].set_ylabel(X_features[3])
ax[2].set_title(r"Z-score normalized")
ax[2].axis('equal')

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
fig.suptitle("distribution of features before, during, after normalization")
plt.show()
```

O gráfico acima mostra a relação entre dois dos parâmetros do conjunto de treinamento, "idade" e "tamanho (sqft)". *Estes são plotados com escala igual*.
- Esquerda: Não normalizado: O intervalo de valores ou a variância da feature 'tamanho (sqft)' é muito maior do que a da idade.
- Meio: O primeiro passo remove a média ou valor médio de cada feature. Isso deixa as features centradas em torno de zero. É difícil ver a diferença para a feature 'idade', mas 'tamanho (sqft)' está claramente em torno de zero.
- Direita: O segundo passo divide pelo desvio padrão. Isso deixa ambas as features centradas em zero com uma escala semelhante.

Vamos normalizar os dados e compará-los com os dados originais.

```python
# normalizar os recursos originais
X_norm, X_mu, X_sigma = zscore_normalize_features(X_train)
print(f"X_mu = {X_mu}, \nX_sigma = {X_sigma}")
print(f"Intervalo pico a pico por coluna em X não normalizado:{np.ptp(X_train,axis=0)}")   
print(f"Intervalo pico a pico por coluna em X normalizado:{np.ptp(X_norm,axis=0)}")
```

O intervalo pico a pico de cada coluna é reduzido de um fator de milhares para um fator de 2-3 pela normalização.

```python
fig, ax = plt.subplots(1, 4, figsize=(12, 3))
for i in range(len(ax)):
    norm_plot(ax[i], X_train[:, i])
    ax[i].set_xlabel(X_features[i])
ax[0].set_ylabel("contagem")
fig.suptitle("Distribuição dos atributos antes da normalização")
plt.show()

fig, ax = plt.subplots(1, 4, figsize=(12, 3))
for i in range(len(ax)):
    norm_plot(ax[i], X_norm[:, i])
    ax[i].set_xlabel(X_features[i])
ax[0].set_ylabel("contagem")
fig.suptitle("Distribuição dos atributos após a normalização")
plt.show()
```

Observe, acima, que o intervalo dos dados normalizados (eixo x) está centrado em torno de zero e é aproximadamente +/- 2. Mais importante ainda, o intervalo é semelhante para cada atributo.

Vamos rodar nosso algoritmo de descida de gradiente novamente com os dados normalizados. Note o **valor muito maior de alpha**. Isso vai acelerar a descida de gradiente.

```python
w_norm, b_norm, hist = run_gradient_descent(X_norm, y_train, 1000, 1.0e-1, )
```

Os recursos escalados obtêm resultados muito mais precisos **muito, muito mais rápido!**. Observe que o gradiente de cada parâmetro é muito pequeno ao final desta corrida bastante curta. Uma taxa de aprendizado de 0.1 é um bom começo para regressão com recursos normalizados.
Vamos plotar nossas previsões versus os valores alvo. Observe que a previsão é feita usando os recursos normalizados, enquanto o gráfico é mostrado usando os valores originais dos recursos.

```python
# Prever o alvo usando recursos normalizados
m = X_norm.shape[0]
yp = np.zeros(m)
for i in range(m):
    yp[i] = np.dot(X_norm[i], w_norm) + b_norm

# plot predictions and targets versus original features    
fig, ax = plt.subplots(1, 4, figsize=(12, 3), sharey=True)
for i in range(len(ax)):
    ax[i].scatter(X_train[:, i], y_train, label='target')
    ax[i].set_xlabel(X_features[i])
    ax[i].scatter(X_train[:, i], yp, color='orange', label='predict')
ax[0].set_ylabel("Price")
ax[0].legend()
fig.suptitle("target versus prediction using z-score normalized model")
plt.show()
```

Os resultados parecem bons. Algumas observações importantes:

- Com múltiplos recursos, não podemos mais ter um único gráfico mostrando os resultados versus os recursos.
- Ao gerar o gráfico, os recursos normalizados foram utilizados. Qualquer previsão feita usando os parâmetros aprendidos de um conjunto de treinamento normalizado também deve ser normalizada.


**Previsão**

O objetivo de gerar nosso modelo é usá-lo para prever os preços de habitação que não estão no conjunto de dados. Vamos prever o preço de uma casa com 1200 pés quadrados, 3 quartos, 1 andar, 40 anos de idade. Lembre-se de normalizar os dados com a média e o desvio padrão derivados quando os dados de treinamento foram normalizados.


```python
# Primeiro, normalize nosso exemplo.
x_house = np.array([1200, 3, 1, 40])
x_house_norm = (x_house - X_mu) / X_sigma
print(x_house_norm)
x_house_predict = np.dot(x_house_norm, w_norm) + b_norm
print(f" preço previsto de uma casa com 1200 pés quadrados, 3 quartos, 1 andar, 40 anos = ${x_house_predict*1000:0.0f}")
```
